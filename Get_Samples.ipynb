```python
import hashlib
import re
import sys
import tarfile
from collections import Counter, defaultdict
from pathlib import Path
import json
import matplotlib.pyplot as plt
import requests
from IPython.display import Image

import nltk
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
nltk.download('all')
```

```python
fname = 'wikitext-103.tar.gz'
url = 'https://dax-cdn.cdn.appdomain.cloud/dax-wikitext-103/1.0.1/' + fname
r = requests.get(url)
Path(fname).write_bytes(r.content)
with tarfile.open(fname) as tar:
    tar.extractall()
train_data = Path('wikitext-103/wiki.train.tokens').read_text(encoding = 'utf8')
```

```python
def flatten_concatenation(matrix):
     flat_list = []
     for row in matrix:
         flat_list += row
     return flat_list
```

```python
heading_pattern = '( \n \n = [^=]*[^=] = \n \n )'
train_split = re.split(heading_pattern, train_data)
train_articles = [x for x in train_split[2::2]]
subb = [re.sub('= = .* = = ', '.', x) for x in train_articles]
split_list = [re.split('(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s', x) for x in subb]
flat_split_list = flatten_concatenation(split_list)
flat_split_list = [re.sub('\n', '', x) for x in flat_split_list]
```

```python
with open('sentences_standard.json', 'w') as file:
    json.dump(flat_split_list, file)
```
